```python
import pandas as pd
```

```python
df=pd.read_csv('orders.csv')
```

```python
# Force all names to lowercase and remove spaces first to be safe
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# Now rename 'list_price' (which was 'List Price') to 'mrp'
df = df.rename(columns={'list_price': 'mrp'})

# CHECK: If this prints 'mrp', we are successful!
print("Current columns:", df.columns.tolist())
```

```python
# 1. Calculate Discount Amount
df['discount'] = df['mrp'] * df['discount_percent'] * 0.01

# 2. Calculate Sale Price (using the new discount column)
df['sale_price'] = df['mrp'] - df['discount']

# 3. Calculate Profit (assuming 'cost_price' was cleaned in Step 1)
df['profit'] = df['sale_price'] - df['cost_price']

# Show the results of our hard work!
df[['mrp', 'discount', 'sale_price', 'profit']].head()
```

```python
df.columns
```

```python

# Wrap the format in quotes and use capital Y for 4-digit years
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
```

```python
print(df['order_date'].dtype)
```

```python
df.columns
```

```python

# droping the following columns:
df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace=True, errors='ignore')
```

```python
#creating data pipeline for python to postgres

engine = sal.create_engine('postgresql://postgres:12345@localhost:5432/postgres')
conn = engine.connect()

# Push the data to your 'df_orders' table
df.to_sql('df_orders', con=conn, index=False, if_exists='append')

print("Success! Your data is now in PostgreSQL.")
```

```python
# Rename columns to match the SQL table structure
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

# Rename the specific columns that might still be different
# (Based on your error log showing 'cost price' and 'discount percent')
df = df.rename(columns={
    'cost_price': 'cost_price', 
    'list_price': 'list_price',
    'discount_percent': 'discount' # Change this if your SQL uses 'discount'
})

# Double check the names
print(df.columns)
```

```python
print(df.columns.tolist())
```

```python
# 1. Force lowercase and underscores again just to be 100% sure
df.columns = [col.lower().replace(' ', '_') for col in df.columns]

# 2. Use 'replace' to create the table structure automatically
with engine.begin() as conn:
    df.to_sql('df_orders', con=conn, index=False, if_exists='replace')
    print("Table structure created and data loaded successfully!")
```
